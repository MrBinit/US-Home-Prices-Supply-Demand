```python
from sklearn.preprocessing import MinMaxScaler

# Create an instance of the normalization scaler
scaler = MinMaxScaler()

# Select the columns to normalize
columns_to_normalize = ['atemp', 'temp', 'hum', 'windspeed']

# Fit the scaler to the selected columns
scaler.fit(day_copy[columns_to_normalize])

# Transform the selected columns using the scaler
normalized_data = scaler.transform(day_copy[columns_to_normalize])

# Replace the original columns with the normalized values
day_copy[columns_to_normalize] = normalized_data
day_copy
```

```python
def seasonname (x):
    if x == 1 : 
        return 'spring'
    elif x ==2:
        return 'summer'
    elif x == 3:
        return 'fall'
    else:
        return 'winter'
```

```python
day_copy['season'] = day_copy['season'].apply(seasonname)

day_copy
```

```python
def year (x):
    if x == 0 : 
        return '2018'
    else:
        return '2019'
day_copy['yr'] = day_copy['yr'].apply(year)
day_copy
```

```python
def weekday (x):
    if x == 0 : 
        return 'sunday'
    elif x == 1:
        return 'monday'
    elif x ==2:
        return 'tuesday'
    elif x == 3:
        return 'wednesday'
    elif x == 4:
        return 'Thrusday'
    elif x == 5:
        return 'Friday'
    else:
        return 'Saturday'
day_copy['weekday'] = day_copy['weekday'].apply(weekday)
day_copy
```

```python
def workingday (x):
    if x == 1 : 
        return 'Working'
    else:
        return 'holiday'
day_copy['workingday'] = day_copy['workingday'].apply(workingday)
day_copy
```

```python
def weathersit (x):
    if x == 1 : 
        return 'Clear'
    elif x == 2:
        return 'blurry'
    elif x ==3:
        return 'Snow'
    else:
        return 'Rainy'
day_copy['weathersit'] = day_copy['weathersit'].apply(weathersit)
day_copy
```

```python
def month_name(x):
    if x == 1:
        return 'January'
    elif x == 2:
        return 'February'
    elif x == 3:
        return 'March'
    elif x == 4:
        return 'April'
    elif x == 5:
        return 'May'
    elif x == 6:
        return 'June'
    elif x == 7:
        return 'July'
    elif x == 8:
        return 'August'
    elif x == 9:
        return 'September'
    elif x == 10:
        return 'October'
    elif x == 11:
        return 'November'
    elif x == 12:
        return 'December'
    else:
        return None

day_copy['mnth'] = day_copy['mnth'].apply(month_name)
day_copy
```

```python
def holiday (x):
    if x == 0 : 
        return 'False'
    elif x == 1:
        return 'True'
    else:
        return None
day_copy['holiday'] = day_copy['holiday'].apply(holiday)
day_copy
```
